```python
import pandas as pd
import numpy as numpy
```

```python
df = pd.read_csv("test.csv")
df['ode_data_list'] = df['ode_data_list'].str.replace('Float32\\[','')
df['ode_data_list'] = df['ode_data_list'].str.replace(']','')
df_new = pd.DataFrame(df.ode_data_list.str.split(';').tolist())
for (columnName, columnData) in df_new.iteritems():
    df_new[columnName] = df_new[columnName].str.strip()
    df_new[columnName] = df_new[columnName].str.split(" ")
for (columnName, columnData) in df_new.iteritems():
    new_cols = [str(columnName)+"_1",str(columnName)+"_2",str(columnName)+"_3",str(columnName)+"_4",str(columnName)+"_5"]
    df_new = df_new.join(pd.DataFrame(df_new.pop(columnName).tolist(), index=df_new.index, columns=new_cols))
df_new.to_csv("data.csv")
df_new
```
